```python
def priceVsFeature(a):
    fig = plt.figure(figsize=(10,3))
    s1=fig.add_subplot(1,2,1)
    s2=fig.add_subplot(1,2,2)


    df=carPrice[a].value_counts()
    df.plot(kind='bar',ax=s1)
    s1.set_title(a+' Histogram')
    s1.set_xlabel(a)


    df = pd.DataFrame(carPrice.groupby([a])['price'].mean().sort_values(ascending = False))
    df.plot(kind='bar',ax=s2)
    plt.title(a+' vs Average Price')
    plt.show()

priceVsFeature('Company')
priceVsFeature('enginetype')
priceVsFeature('fueltype')
priceVsFeature('fuelsystem')
priceVsFeature('enginelocation')
priceVsFeature('carbody')
priceVsFeature('doornumber')
priceVsFeature('cylindernumber')
priceVsFeature('drivewheel')
```

```python
def scatter(x,fig):
    plt.subplot(2,2,fig)
    plt.scatter(carPrice[x],carPrice['price'])
    plt.title(x+' vs Price')
    plt.ylabel('Price')
    plt.xlabel(x)

plt.figure(figsize=(10,8))

scatter('carlength', 1)
scatter('carwidth', 2)
scatter('carheight', 3)
scatter('curbweight', 4)

plt.tight_layout()
plt.show()
```

Inference : 
- Car price has +ve correlation with feature : car width, carl ength and curb weight.
- Car height seems to have no specific relation with car price.

```python
def pair_plot(x,y,z):
    sb.pairplot(carPrice, x_vars=[x,y,z], y_vars='price',size=4, aspect=1, kind='scatter')
    plt.show()

pair_plot('wheelbase', 'citympg', 'highwaympg')
pair_plot('enginesize', 'boreratio', 'stroke')
pair_plot('compressionratio', 'horsepower', 'peakrpm')
```
